## <b>Machine Learning

Titanic

```python
import pandas as pd
df= pd.read_csv("/content/drive/MyDrive/datasets/titanic_train.csv")

X= df["Fare"].copy()
X= X.values.reshape(-1, 1)
y= df["Survived"].copy()
```

```python
from sklearn.linear_model import LogisticRegression
clf= LogisticRegression()
clf.fit(X, y)
```

```python
clf.score(X, y)
```

<b> Probabilidade de sobrevivência 0 = Negativa, 1 = Positiva

```python
import numpy as np
clf.predict(np.array([[50]]))
```

```python
clf.predict(np.array([[70]]))
```

<b>O primeiro resultado indica a probabilidade de ocorrer a classe 0 (não sobrevivência), o segundo indica a classe 1 (sobrevivência)

```python
clf.predict_proba(np.array([[50]]))
```

```python
clf.predict_proba(np.array([[70]]))
```

<b>Recursive Feature Elimination

```python
from sklearn.feature_selection import RFECV

selector= RFECV(estimator= LogisticRegression())

selector.fit(X_treino, y_treino)
```

```python
selector.n_features_
```

```python
selector.get_support(indices= True)
```

```python
cols_indices= selector.get_support(indices= True)
cols= X_treino.columns[cols_indices]
```

```python
X_treino= X_treino[cols]
X_teste= X_teste[cols]
```

```python
clf.fit(X_treino, y_treino)
```

```python
clf.score(X_teste, y_teste)
```
