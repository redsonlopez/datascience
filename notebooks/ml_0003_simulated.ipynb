Non-Linear Regression with Gauss-Newton Method

```python
import numpy as np
from scipy.optimize import curve_fit
import pylab

# Model definition
def func(x, p1, p2):
  return p1 * np.exp(p2 * xdata)

# Data set
xdata = np.linspace(0, 4, 50)
y = func(xdata, 1.3, 0.5)
ydata = y + 0.2 * np.random.normal(size=len(xdata))

# Call curve_fit
popt, pcov= curve_fit(func, xdata, ydata, p0= (10.0, -1.0))
p1, p2= popt
# Printing of optimal parameters
print("Optimal parameters are p1=%g, p2=%g"%(p1, p2))

# Plot of results
yfitted= func(xdata, *popt)
pylab.plot(xdata, ydata, "o", label= "dados $y_i$")
pylab.plot(xdata, yfitted, "-", label= "fit $f(x_i)$")
pylab.xlabel("x")
pylab.ylabel("y")
pylab.legend()

ax = pylab.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

pylab.show()
```
