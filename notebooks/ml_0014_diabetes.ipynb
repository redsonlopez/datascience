Data preview:

```python
import pandas as pd
df= pd.read_csv("/content/drive/MyDrive/database/diabetes.csv")
df.head(3)
```

Separation of training and testing data:

```python
X = pd.get_dummies(df.iloc[:, :8]).values
y= df["diabetes"].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.20, random_state= 82)
```

Model training and testing:

```python
from sklearn import linear_model
regr= linear_model.LinearRegression()
regr.fit(X_train, y_train)
y_pred= regr.predict(X_test)
```

Metrics for calculating model performance:

```python
from sklearn.metrics import mean_squared_error, r2_score
print("Mean squared error: %.2f"%mean_squared_error(y_test, y_pred))
print("Forecast coefficient: %.2f" %r2_score(y_test, y_pred)) # Forecast coefficient. 1 is perfect
```
