```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFECV

class Model:
  def __init__(self, df_treino: pd.DataFrame,
               df_teste: pd.DataFrame) -> None:
    self.df_treino= df_treino
    self.df_teste= df_teste
    self.cfl= LogisticRegression()

  def separar_variaveis(self) -> None:
    self.X_treino= self.df_treino.drop(columns= "Reached.om.time_Y.N")
    self.y_treino= self.df_treino["Reached.om.time_Y.N"]

    self.X_teste= self.df_teste.drop(columns= "Reached.om.time_Y.N")
    self.y_teste= self.df_teste["Reached.om.time_Y.N"]

  def selecionar_variaveis(self) -> None:
    seletor= RFECV(self.clf)
    seletor.fit(self.X_treino, self.y_treino)

    cols_indices= seletor.getsupport(indices= True)
    cols= self.X_treino.columns[cols_indices]

    self.X_treino= self.X_treino[cols]
    self.X_teste= self.X_teste[cols]

  def fit(self):
    self.separar_variaveis()
    self.selecionar_variaveis()
    self.clf.fit(self.X_treino, self.y_treino)

  def predict(self, X: pd.DataFrame):
    predictions= self.clf.predict(X)
    return predictions

  def score(self):
    return self.clf.score(self.X_teste, self.y_teste)
```
