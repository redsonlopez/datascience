Bayesian Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import BayesianRidge, LinearRegression

# Generation of simulated data with random weights
np.random.seed(0)
n_samples, n_features= 100, 100
X= np.random.randn(n_samples, n_features)

# Creating weights with lambda= 4
lambda_= 4.0
w= np.zeros(n_features)

# Using only 10 weights
relevant_features= np.random.randint(0, n_features, 10)
for i in relevant_features:
  w[i]= stats.norm.rvs(loc= 0, scale= 1.0/ np.sqrt(lambda_))

# Alpha Precision= 50
alpha_= 50.0
noise= stats.norm.rvs(loc= 0, scale= 1.0/ np.sqrt(alpha_), size= n_samples)

# Setup for output
y= np.dot(X, w)+ noise

# Model fit and comparison with OLS(ordinary least squares)
clf= BayesianRidge(compute_score= True)
clf.fit(X, y)
ols= LinearRegression()
ols.fit(X, y)
```

```python
# Implementation for generating the graph
lw= 2
plt.figure(figsize= (6, 5))
plt.title("Model weights")
plt.plot(clf.coef_, color= "lightgreen", linewidth= lw, label= "Bayesian Linear Regression")
plt.xlabel("Features")
plt.ylabel("Weight values")
plt.legend(prop= dict(size= 10))
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
plt.show()
```
