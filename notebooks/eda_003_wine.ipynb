```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 5)
df= pd.read_csv("/content/drive/MyDrive/datasets/winemag.csv", index_col= 0)
df.head(3)
```

```python
df.shape
```

```python
df.price.describe()
```

```python
df.taster_name.value_counts()
```

```python
df_points_mean = df.points.mean()
df.points.map(lambda p: p - df_points_mean)
```

```python
def remean_points(row):
    row.points = row.points - df_points_mean
    return row

df.apply(remean_points, axis='columns')
```

```python
df.head(3)
```

```python
df_points_mean = df.points.mean() # These operators are faster than map() or apply(), However, they are not as flexible as map() or apply()
df.points - df_points_mean
```

```python
df.country + " - " + df.region_1
```

```python
bargain_idx = (df.points / df.price).idxmax()
bargain_wine = df.loc[bargain_idx, "title"]

bargain_wine # Title of the wine with the highest points-to-price ratio in the dataset
```

```python
n_trop = df.description.map(lambda desc: "tropical" in desc).sum()
n_fruity = df.description.map(lambda desc: "fruity" in desc).sum()
descriptor_counts = pd.Series([n_trop, n_fruity], index=['tropical', 'fruity']) # Checking the string 'tropical', 'fruity' and counting the number of times this is true with series combining the two values
```

```python
def stars(row):
    if row.country == 'Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1

star_ratings = df.apply(stars, axis='columns') # Series with number of stars corresponding to each rating in the dataset
```

```python
df.groupby('points').points.count()
```

```python
df.groupby('points').price.min()
```

```python
df.groupby('winery').apply(lambda df: df.title.iloc[0]) # Name of the first wine reviewed from each winery in the dataset
```

```python
df.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()]) # Best wine by country and province
```

```python
df.groupby(['country']).price.agg([len, min, max]) # Simple statistical summary of the dataset
```

```python
countries_reviewed = df.groupby(['country', 'province']).description.agg([len]) # Multi-index
countries_reviewed
```

```python
countries_reviewed.reset_index() # Converting back to a regular index
```

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
countries_reviewed.sort_values(by=['country', 'len']) # Sort by more than one column at a time
```

```python
best_rating_per_price = df.groupby('price')['points'].max().sort_index()
best_rating_per_price # What is the best wine I can buy for a given amount of money?
```

```python
price_extremes = df.groupby('variety').price.agg([min, max])
price_extremes
```

```python
sorted_varieties = price_extremes.sort_values(by=['min', 'max'], ascending=False)
sorted_varieties
```

```python
reviewer_mean_ratings = df.groupby('taster_name').points.mean()
reviewer_mean_ratings
```

```python
country_variety_counts = df.groupby(['country', 'variety']).size().sort_values(ascending=False) # Multi-index
country_variety_counts
```
