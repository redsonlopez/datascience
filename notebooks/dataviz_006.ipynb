```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Configurações para criar legendas ao final das linhas
%matplotlib inline
fig = plt.figure(figsize=(10,5)) # Largura e altura do gráfico
ax = fig.add_subplot(111)
ax.text(11.5, 177, 'Linha 1', ha='left', va='top', fontsize=12,color='gray')
ax.text(11.5, 140, 'Linha 2', ha='left', va='top', fontsize=12,color='blue')

# Dados para o exemplo
x= ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
y1= [160,184,241,149,180,161,132,202,160,139,149,177]
y2= [160,184,237,148,181,150,123,156,126,104,124,140]

# Obtém as coordenadas finais de cada linha
line1_coords = [(xi, yi) for xi, yi in zip(x, y1)]
line2_coords = [(xi, yi) for xi, yi in zip(x, y2)]

# Adiciona as legendas nos pontos de cada linha
for coord in line1_coords:
    plt.text(coord[0], coord[1], str(coord[1]), ha='right', va='bottom', color= 'lightgray')
for coord in line2_coords:
    plt.text(coord[0], coord[1], str(coord[1]), ha='right', va='bottom', color= 'lightgray')

# Cria o gráfico de linhas usando o Seaborn
sns.lineplot(x=x, y=y1, marker= 'o', color='gray', ax=ax)
sns.lineplot(x=x, y=y2, marker= 's', color='blue', ax=ax)

plt.xlim() # Domínio de x
plt.ylim(0, 300) # Ímagem de y

# Estilo e Tamanho da Fonte
sns.set_style('ticks')
plt.rcParams.update({'font.size': 8})

plt.suptitle('Por favor, aceite a solicitação',fontsize=20,ha='right',va='top',color='#000088', y=1) # Titulo
plt.ylabel('Quantidade Produzida', color='gray') # Título eixo y

# Anotações no gráfico
plt.annotate("Esse gráfico foi feito\nsem saber o que está sendo feito,\npor alguém que não sabe o que foi feito.",
             xy=(2,0),xytext=(8,275),color='#616161',fontsize=10)

plt.grid(False) # Grade

sns.despine()
```
