<b>Filtro para 'Receita' e 'Despesa'

```python
df1.CLASSIFICAÇÃO.unique()
```

```python
valores_receita= ['RECEITAS OPERACIONAIS', 'RECEITAS FINANCEIRAS',
                  'RECEITAS NÃO OPERACIONAIS', 'DEDUÇÕES DA RECEITA']
valores_despesa= ['CUSTOS VARIÁVEIS', 'DESPESAS FIXAS',
                  'DESPESAS FINANCEIRAS', 'INVESTIMENTOS',
                  'DESPESAS NÃO OPERACIONAIS', 'PARTICIPAÇÕES']

receita= df1.loc[df1["CLASSIFICAÇÃO"].isin(valores_receita)]
despesa= df1.loc[df1["CLASSIFICAÇÃO"].isin(valores_despesa)]
```

<b>Lucro líquido

```python
receita['VALOR'].sum() - despesa["VALOR"].sum()
```

<b>Gráfico de barras para 'Receita' por 'Unidade'
