```python
class Methods:
  def metodo_de_instancia(self, x):
    print(self, x)

  @staticmethod
  def metodo_estatico(x):
    print(x)

  @classmethod
  def metodo_de_classe(cls, x):
    print(cls, x)
```

```python
m= Methods()

m.metodo_de_instancia("Chamando um metodo de instância")
```

```python
m.metodo_de_classe("Chamando um metodo de classe")
```

```python
#@title Card

class Card:
  def __init__(self, limite):
    self.limite= limite

  @classmethod
  def gold(cls):
    return cls(1000)

  @classmethod
  def platinum(cls):
    return cls(3000)

  @classmethod
  def diamond(cls):
    return cls(8000)
```

```python
card_gold= Card.gold()
print(card_gold.limite)
```

```python
from datetime import datetime

class VendasDiarias:
  '''
  Analisa as vendas realizadas na data de referência.
  '''
  def __init__(self, data):
    self.data= VendasDiarias.converter_data(data)

  @staticmethod
  def converter_data(data):
    return datetime.strptime(data, '%d/%m/%y')
```

```python
vendas= VendasDiarias('03/07/21')
print(vendas.data)
```

```python
print(type(vendas.data))
```

```python
#@title Cartão

class Cartao:
  def __init__(self):
    self._limite= 100

  @property
  def limite(self):
    print("Property getter")
    return self._limite

  @limite.setter
  def limite(self, limite):
    print("Property setter")
    if limite < 0:
      raise ValueError("O limite não pode ser negativo")
    else:
      self._limite= limite
```

```python
cartao= Cartao()
cartao.limite= 150
```

```python
cartao.limite
```
