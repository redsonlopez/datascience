```python
#@title conta

class Conta:
  """
  Classe principal para conta bancaria
  """
  def __init__(self, numero_da_conta, saldo):
    self.numero_da_conta= numero_da_conta
    self.saldo= saldo

  def depositar(self, valor):
    self.saldo += valor

  def sacar(self, valor):
    self.saldo -= valor
```

```python
#@title Conta Poupança

class ContaPoupanca(Conta):
  def calcular_rendimento(self):
    self.saldo += self.saldo * 0.005
```

```python
cp= ContaPoupanca(123456, 200)
```

```python
cp.depositar(100)
cp.saldo
```

```python
cp.sacar(50)
cp.saldo
```

```python
cp.calcular_rendimento()
cp.saldo
```

```python
from abc import ABC, abstractmethod

class Conta(ABC):
  def __init__(self, numero_da_conta, saldo):
    self.numero_da_conta= numero_da_conta
    self.saldo= saldo

  def depositar(self, valor):
    self.saldo += valor

  @abstractmethod
  def sacar(self, valor):
    raise NotImplementedError
```

```python
class ContaPoupanca(Conta):
  def sacar(self, valor):
    if valor <= self.saldo:
      self.saldo -= valor
    else:
      print("A operação não pode ser realizada.")

  def calcular_rendimento(self):
    self.saldo += self.saldo * 0.005
```

```python
cp= ContaPoupanca(123456, 200)
```

```python
class ContaCorrente(Conta):
  def __init__(self, numero_da_conta, saldo, limite):
    super().__init__(numero_da_conta, saldo)
    self.limite= limite

  def sacar(self, valor):
    if valor <= self.saldo + self.limite:
      self.saldo -= valor
    else:
      print("A operação não pode ser realizada.")
```
