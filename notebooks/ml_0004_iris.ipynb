```python
import pandas as pd

dataset= pd.read_csv("https://raw.githubusercontent.com/redsonlopez/datasets/main/iris.xls")
dataset.head(3)
```

```python
import numpy as np
import PIL.Image
import urllib.request
import matplotlib.pyplot as plt
image_data = np.array(PIL.Image.open(urllib.request.urlopen("https://raw.githubusercontent.com/redsonlopez/datasets/main/iris_image.jpg")))
plt.axis("off")
plt.imshow(image_data);
```

```python
X= dataset.iloc[:, :4].values
y= dataset["Species"].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.20, random_state= 82)

from sklearn.svm import SVC
linear_svc= SVC(kernel="linear").fit(X_train, y_train)
prediction= linear_svc.predict(X_test)

from sklearn import metrics
print("SVC Accuracy:", "{:.2f} %".format((metrics.accuracy_score(prediction, y_test)* 100)))

Poly_svc= SVC(kernel= "poly", degree= 4).fit(X_train, y_train)
prediction= Poly_svc.predict(X_test)
print("Polynomial SVC Accuracy:","{:.2f} %".format((metrics.accuracy_score(prediction, y_test)*100)))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
cm = confusion_matrix(y_test, prediction)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.xticks(())
plt.yticks(())
plt.show()
```
