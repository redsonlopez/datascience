<i>Cluster</i>

```python
X = [25, 36, 89, 77, 72, 65, 27, 45, 35, 20, 25, 44, 26, 84, 29, 28, 85, 82]
y = [7204, 2277, 4606, 1425, 5813, 5693, 3061, 7371, 3095, 4831, 5411, 4371, 6308, 4877, 6674, 2662, 7139, 5584]

import numpy as np
base_salary= np.array([[25, 7204], [36, 2277], [89, 4606], [77, 1425], [72, 5813], [65, 5693], [27, 3061], [45, 7371],
 [35, 3095], [20, 4831], [25, 5411], [44, 4371], [26, 6308], [84, 4877], [29, 6674], [28, 2662], [85, 7139], [82, 5584]])

import seaborn as sns
sns.relplot(x= X, y= y);
```

```python
print(f"The number of existing records in the database = {base_salary.shape[0]}")
```

n_clusters formula: <i>&radic;n/2</i><br>
where <i>n</i> is the number of existing records in the database;<br>
<i>&radic;18/2 = &radic;9 = 3<i/>

```python
from sklearn.preprocessing import StandardScaler
scaler_salary= StandardScaler()
base_salary= scaler_salary.fit_transform(base_salary)

from sklearn.cluster import KMeans
kmeans_salary= KMeans(n_clusters= 3, n_init= 10)
kmeans_salary.fit(base_salary)
```

```python
centroids= kmeans_salary.cluster_centers_
scaler_salary.inverse_transform(kmeans_salary.cluster_centers_)
labels= kmeans_salary.labels_

sns.relplot(x= base_salary[:, 0], y= base_salary[:, 1], hue= labels);
```
