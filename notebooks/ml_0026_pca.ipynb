```python
import pandas as pd
df= pd.read_csv("https://raw.githubusercontent.com/redsonlopez/datasets/main/iris.xls")
X= df.iloc[:, 1:5]

X.head(3)
```

```python
from sklearn.decomposition import PCA
pca= PCA(n_components= 4)
pca.fit(X)

print(pca.explained_variance_ratio_)
```

```python
import seaborn as sns
sns.set(rc={"axes.facecolor":"5B5B5B", "figure.facecolor":"5B5B5B"})

sns.barplot(x= X.columns, y= pca.explained_variance_ratio_, palette= "Blues")
sns.despine()
```
