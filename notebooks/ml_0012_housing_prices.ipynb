Modelo de <i>Regressão</i><br><br>
Pré-visualização dos dados:

```python
import pandas as pd
df= pd.read_csv("https://raw.githubusercontent.com/redsonlopez/ds/main/datasets/housing_prices_train.csv")
df.head(3)
```

Separação dos dados de treino e teste:

```python
from sklearn.model_selection import train_test_split
features= ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X= df[features]
y= df.SalePrice
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size= 0.20, random_state=1)
```

Treino e teste do modelo:

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor()
rf.fit(X_train, y_train)
predicted = rf.predict(X_val)
```

Métricas para calcular o desempenho do modelo:

```python
from sklearn.metrics import mean_absolute_error, r2_score
print("Mean absolute error: %.0f"%mean_absolute_error(y_val, predicted)) # Quanto menor o valor do erro, melhor.
print("Forecast coefficient: %.2f" %r2_score(y_val, predicted)) # Capacidade de capturar a variação nos valores reais da variável de destino.
```
