```python
#@title Carteira

class Carteira:
  """
  Representa uma carteira digital para armazenar e manipular saldo monetário
  """
  def __init__(self, limite):
    self.saldo= 0.0
    self.limite= limite

  def adicionar_fundos(self, valor):
    """
    Adiciona o valor indicado no saldo da carteira
    """
    if valor < 0:
      raise ValueError("Não é possível adicionar um valor negativo na carteira.")
    else:
      self.saldo += valor
    print(f"Operação realizada com sucesso!\nSaldo atual R${self.saldo}")

  def remover_fundos(self, valor):
    """
    Remove o valor indicado no saldo da carteira
    caso o saldo seja suficiente
    """
    if valor < 0:
      raise ValueError("Não é possível remover um valor negatico da carteira.")
    if self.saldo >= valor:
      self.saldo -= valor
      print(f"Operação realizada com sucesso!\nSaldo atual R${self.saldo}")
    else:
      print("Operação não realizada. Saldo insuficiente.")

  def exibir_saldo(self):
    """
    Exibe o saldo atual da carteira
    """
    print(f"Seu saldo é de R${self.saldo}")

c1= Carteira(0)
c2= Carteira(100)
```

```python
c1.adicionar_fundos(50)
```

```python
c1.remover_fundos(30)
```

```python
c1.exibir_saldo()
```

```python
c2.exibir_saldo()
```

```python
c2.adicionar_fundos(50)
```

Desenvolvendo uma classe abstrata

```python
#@title Cartão Base
from abc import ABC, abstractmethod

class CartaoBase(ABC):
  """
  Classe abstrata que representa um cartão de credito
  """
  @abstractmethod
  def calcular_cash_back(self, valor):
    pass
```

```python
#@title Desenvolvendo classes concretas

class CartaoGold(CartaoBase):
  def calcular_cash_back(self, valor):
    return valor * 0.03

class CartaoPlatinum(CartaoBase):
  def calcular_cash_back(self, valor):
    return valor * 0.05

class CartaoDiamond(CartaoBase):
  def calcular_cash_back(self, valor):
    return valor * 0.08
```

```python
# Exemplos de instância para cada classe

print(CartaoGold().calcular_cash_back(100))
print(CartaoPlatinum().calcular_cash_back(100))
print(CartaoDiamond().calcular_cash_back(100))
```

```python
#@title Cartão

class Cartao:
  def __init__(self):
    self._limite= 100

  def _exibir_limite(self):
    print("Chamando metodo exibir_limite")
    return self._limite

  def _alterar_limite(self, novo_limite):
    print("Chamando metodo alterar_limite")
    if novo_limite < 0:
      raise ValueError("O limite não pode ser negativo.")
    else:
      self._limite= novo_limite

  limite= property(_exibir_limite, _alterar_limite)
```

```python
card= Cartao()
card.limite= 150
```
