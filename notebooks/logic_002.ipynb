```python
#Esse eu preciso entender:

menu_1 = ["Sardine", "Banana", "Wine", "Cookie",]
menu_2 = ["Sardine", "Banana", "Wine", "Cookie", "Cookie", "Ice Cream"]

def menu_is_boring(meals):
    """Given a list of meals served over some period of time, return True if the
    same meal has ever been served two days in a row, and False otherwise.
    """

    for eat in range(len(meals)-1):
        if meals[eat] == meals[eat+1]:
            return True
    return False


menu_is_boring(menu_2)
```
