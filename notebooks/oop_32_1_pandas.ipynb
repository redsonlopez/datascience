### Projeto de análise de dados

```python
df_lojas= pd.DataFrame({
    "id_loja": [1, 2, 3, 4, 5],
    "tipo_loja": ["supermercado", "hipermercado", "hipermercado", "supermercado", "supermercado"],
    "UF": ["MG", "ES", "MG", "MG", "ES"]
})

df_lojas.to_csv("lojas.csv", index= False)
df_lojas
```

```python
df_produtos= pd.DataFrame({
    "id_produto": [10, 11, 12, 13, 14, 15],
    "categoria": ["alimentos", "bebidas", "limpeza", "alimentos", "limpeza", "bebidas"]
})

df_produtos.to_csv("produtos.csv", index= False)
df_produtos
```

```python
df_vendas= pd.DataFrame({
    "id_venda": np.arange(1, 21),
    "id_loja": np.random.randint(1, 6, 20),
    "id_produto": np.random.randint(10, 15, 20),
    "receita": np.random.randint(100, 1000, 20)
})

df_vendas.to_csv("vendas.csv", index= False)
df_vendas.head()
```

```python
class Vendas:
  def __init__(self):
    self.df_lojas= pd.read_csv("lojas.csv")
    self.df_produtos= pd.read_csv("produtos.csv")
    self.df_vendas= pd.read_csv("vendas.csv")

  def combinar_bases(self):
    self.df_vendas= self.df_vendas.merge(self.df_produtos)
    self.df_vendas= self.df_vendas.merge(self.df_lojas)

  def calcular_vendas_por_tipo_loja(self):
    df= self.df_vendas.groupby("tipo_loja")["receita"].sum()
    return df

  def calcular_vendas_por_categoria(self):
    df= self.df_vendas.groupby("categoria")["receita"].sum()
    return df
```
