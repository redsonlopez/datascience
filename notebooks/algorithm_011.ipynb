```python
class carro():
  def __init__(self, modelo, cor):
    self.modelo = modelo
    self.cor = cor

  def mostrar(self):
      print("Modelo é", self.modelo)
      print("Cor é", self.cor)

audi = carro("audi a4", "azul")
ferrari = carro("ferrari488", "verde")
audi.mostrar()
ferrari.mostrar()
```
