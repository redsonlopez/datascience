```python
from sklearn import datasets
iris = datasets.load_iris()
selected_keys = ['data', 'target_names', 'feature_names', 'target', 'DESCR']
selected_data = {key: iris[key] for key in selected_keys}

X = selected_data["data"][:, 3:]  # largura da pétala
y = (selected_data["target"] == 2).astype(int)  # 1 se for Iris-Virginica, caso contrário, 0

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X, y)

import numpy as np
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)

import matplotlib.pyplot as plt
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")

plt.xlabel("Petal Width")
plt.ylabel("Probability")

plt.legend(loc= "right")
plt.show()
```
