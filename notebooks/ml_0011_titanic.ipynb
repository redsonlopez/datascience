## <b>Machine Learning</b> <i>Titanic</i>

```python
import numpy as np
import pandas as pd

train_csv = '/content/drive/MyDrive/database/titanic_train.csv'
train = pd.read_csv(train_csv)

test_csv = '/content/drive/MyDrive/database/titanic_test.csv'
test = pd.read_csv(test_csv)
```

```python
print(f"The dataset has {train.shape[0]} rows and {train.shape[1]} columns. \n")
train.tail(3)
```

```python
print(f"The dataset has {test.shape[0]} rows and {test.shape[1]} columns. \n")
test.head(3)
```

```python
women = train.loc[train.Sex == "female"]["Survived"]
rate_women = round(sum(women)/len(women)* 100, 1)
print(f"% of women who survived: {rate_women}")
```

```python
men = train.loc[train.Sex == "male"]["Survived"]
rate_men = round(sum(men)/len(men)* 100, 1)
print(f"% of men who survived: {rate_men}")
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train[features])
X_test = pd.get_dummies(test[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test.PassengerId, 'Name': test.Name, 'Survived': predictions})
```

```python
from sklearn.model_selection import cross_val_score

# Concatenating the training and testing sets
all_data = pd.concat([train, test])

# Creating the input variables (features)
X_all = pd.get_dummies(all_data[features])

# Separating back into training and testing set
X_train = X_all[:len(train)]
X_test = X_all[len(train):]

# Creating the true labels for the training set
y_train = train["Survived"]

# Creating the model
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)

# Applying cross validation
scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')

# Calculating the mean and standard deviation of scores
mean_accuracy = round(scores.mean() *100, 2)
std_accuracy = round(scores.std() * 100, 2)

print("Average accuracy of cross validation:", mean_accuracy, "%")
print("Standard deviation of cross-validation accuracy:", std_accuracy, "%")
```
