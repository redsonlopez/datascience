## <b><i>Dados de um Restaurante</b></i>

<i>Conjunto de dados tips.</i><br><br>

<b>Este dataset contém as seguintes informações:</b><br><br><i>
**total_bill:** valor total da conta.<br>
**tip:** valor da gorjeta.<br>
**sex:** sexo da pessoa que pagou a conta.<br>
**smoker:** indica se a pessoa que pagou a conta é fumante.<br>
**day:** dia da semana<br>
**time:** horário da refeição: almoço (lunch) ou jantar (dinner).<br>
**size:** indica o número de pessoas que estavam na mesa.<br>

```python
import seaborn as sns

df= sns.load_dataset("tips")

df.head(3) #Visualização das colunas e dos 3 primeiros registros
```

```python
print(f"Quantidade de registros no conjunto de dados: {df.shape[0]}\n")
print(f"Frequencia de resultados unicos para cada coluna:\n{df.nunique()}\n")
print("Dias de funcionamento do estabelecimento:\n", df['day'].unique())
```

<b>Relação entre o valor da gorjeta e o dia da semana

```python
ax= sns.barplot(data= df, x="day", y= "tip")
```

<b>Valor médio da gorjeta por sexo

```python
ax= sns.barplot(data= df, x= "sex", y= "tip")
```

<b>Valor médio da gorjeta em cada horário da refeição

```python
ax= sns.barplot(data= df, x= "time", y= "tip")
```

<b>Relação entre o valor total da conta e o valor da gorjeta

```python
ax= sns.scatterplot(data= df, x= "total_bill", y= "tip")
```

<b>Linearidade entre o valor total da conta e o valor da gorjeta

```python
ax= sns.regplot(data= df, x= "total_bill", y= "tip")
```

<b>Distribuição entre os valores da conta

```python
ax= sns.histplot(data= df, x="total_bill")
```

<b>Distribuição dos valores das gorjetas

```python
ax= sns.histplot(data= df, x="tip")
```
