```python
import pandas as pd
import numpy as np
import xlwt

# Criação de dados de exemplo
dados = {
    'Feature1': np.round(np.random.uniform(-50.0, 50.0, size=100), 1),
    'Feature2': np.round(np.random.uniform(-50.0, 50.0, size=100), 1)
}

# Criação do DataFrame usando os dados
df = pd.DataFrame(dados)

# Criação de um objeto de workbook do xlwt
workbook = xlwt.Workbook()

# Criação de uma planilha no workbook
sheet = workbook.add_sheet('Sheet1')

# Escrever os dados na planilha
for i, col_name in enumerate(df.columns):
    for j, value in enumerate(df[col_name]):
        sheet.write(j, i, value)  # Escrever os valores nas células (sem a primeira linha)

# Salvando o workbook como um arquivo XLS
workbook.save('/content/drive/MyDrive/database/data_dumb.xls')

df.head()
```
