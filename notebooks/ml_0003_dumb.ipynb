Non-Linear Regression with Gauss-Newton Method

```python
import numpy as np
from scipy.optimize import curve_fit
import pylab
import xlrd

wb= xlrd.open_workbook("/content/drive/MyDrive/datasets/data_dumb.xls")
sh1= wb.sheet_by_name(u"Sheet1")
xdata= np.array(sh1.col_values(0))
ydata= np.array(sh1.col_values(1))

# Model definition
def func(x, p1, p2):
  return p1 * np.exp(p2 * xdata)

# Call curve_fit
popt, pcov= curve_fit(func, xdata, ydata, p0= (10.0, -1.0))
p1, p2= popt
# Printing of optimal parameters
print("Optimal parameters are p1=%g, p2=%g"%(p1, p2))

# Plot of results
yfitted= func(xdata, *popt)
pylab.plot(xdata, ydata, "o", label= "dados $y_i$")
pylab.plot(xdata, yfitted, "-", label= "fit $f(x_i)$")
pylab.xlabel("x")
pylab.ylabel("y")
pylab.legend()

ax = pylab.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

pylab.show()
```
