```python
import pandas as pd

dados = {
    'turma':['A', 'B', 'C'],
    'qtde_alunos':[33, 50, 45]
    }
df = pd.DataFrame(dados)

df

df.plot.bar(x = "turma", y = "qtde_alunos")

df.plot.barh(x = "turma", y = "qtde_alunos")

df.plot.line(x = "turma", y = "qtde_alunos")

df.set_index("turma").plot.pie(y = "qtde_alunos")
```
