```python
import pandas as pd
import folium

saude_fidedigna= pd.read_csv("~/Downloads/datacic/unidades/Saúde - Fidedigna.csv", index_col= None)
atributos= ["REGIONAL", "UNIDADE DE SAÚDE", "LOGRADOURO POR GEORREFERENCIAMENTO", 
            "NOME DO LOGRADOURO POR GEORREFERENCIAMENTO", "NÚMERO POR GEORREFERENCIAMENTO", "BAIRRO",
            "COORDENADAS GEOGRÁFICAS POR GEORREFERENCIAMENTO"]
fidedigna= saude_fidedigna[atributos]
fidedigna.shape
```

```python
fidedigna.loc[(fidedigna["NOME DO LOGRADOURO POR GEORREFERENCIAMENTO"] == "ANDIROBA") &\
(fidedigna["NÚMERO POR GEORREFERENCIAMENTO"] == "100") & (fidedigna["BAIRRO"] == "SAO PAULO"),\
"COORDENADAS GEOGRÁFICAS POR GEORREFERENCIAMENTO"] = "-19.865590, -43.926254"

fidedigna.loc[fidedigna["COORDENADAS GEOGRÁFICAS POR GEORREFERENCIAMENTO"] == "--19.83279347, -43.94601295",\
"COORDENADAS GEOGRÁFICAS POR GEORREFERENCIAMENTO"] = "-19.83279347, -43.94601295"

fidedigna[['Latitude', 'Longitude']] = fidedigna['COORDENADAS GEOGRÁFICAS POR GEORREFERENCIAMENTO']\
.str.rsplit(',', n=1, expand=True).astype(float).copy()
fidedigna.shape
```

```python
mymap = folium.Map(location=[fidedigna['Latitude'].mean(), fidedigna['Longitude'].mean()], zoom_start=11.5)
for index, row in fidedigna.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup= "{}, {}, {}, {}"
                  .format(row['LOGRADOURO POR GEORREFERENCIAMENTO'], row['NOME DO LOGRADOURO POR GEORREFERENCIAMENTO'],
                          row['NÚMERO POR GEORREFERENCIAMENTO'], row["BAIRRO"])).add_to(mymap)

folium.Marker([-19.923776, -43.936434], popup= "PBH", icon= folium.Icon(color= "darkred", icon= "info-sign")).add_to(mymap)
mymap
```

```python
mymap.save('mapa_saude.html')
```
