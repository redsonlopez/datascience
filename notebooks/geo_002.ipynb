```python
import pandas as pd
import folium

df= pd.read_csv("~/Downloads/Saúde - Fidedigna.csv")
print(f"Linhas= {df.shape[0]}, Colunas= {df.shape[1]}")
df[df["NOME DO LOGRADOURO POR GEORREFERENCIAMENTO"] == "SALINAS"]
```

```python
atributos= ["FOI REALIZADA VISITA?", "REGIONAL", "UNIDADE DE SAÚDE", "LOGRADOURO POR GEORREFERENCIAMENTO", 
            "NOME DO LOGRADOURO POR GEORREFERENCIAMENTO", "NÚMERO POR GEORREFERENCIAMENTO", "BAIRRO",
            "COORDENADAS GEOGRÁFICAS POR GEORREFERENCIAMENTO"]
df= df[atributos]
df= df[df["FOI REALIZADA VISITA?"] == "NÃO"]
print(f"Linhas= {df.shape[0]}, Colunas= {df.shape[1]}")
```

```python
df[['Latitude', 'Longitude']] = df['COORDENADAS GEOGRÁFICAS POR GEORREFERENCIAMENTO'].str.rsplit(',', n=1, expand=True).astype(float)
display(df.head(2))
df.shape
```

```python
# Criar um mapa
mymap = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=12)

folium.Marker([-19.9245, -43.9365], popup= "GCUST", icon= folium.Icon(color= "darkred", icon= 'info-sign')).add_to(mymap)
folium.Marker([-19.9803, -44.0574], popup= "Hedson", icon= folium.Icon(color= "gray", icon= "info-sign")).add_to(mymap)
folium.Marker([-19.8444, -43.9391], popup= "JP2", icon= folium.Icon(color= "gray", icon= "info-sign")).add_to(mymap)
folium.Marker([-19.9057, -44.0125], popup= "Thiago", icon= folium.Icon(color= "gray", icon= "info-sign")).add_to(mymap)
folium.Marker([-19.8718, -44.0086], popup= "Eros", icon= folium.Icon(color= "gray", icon= "info-sign")).add_to(mymap)

# Adicionar marcadores ao mapa
for index, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup= "{}, {}, {}, {}, {}"
                  .format(row['UNIDADE DE SAÚDE'], row['LOGRADOURO POR GEORREFERENCIAMENTO'], row['NOME DO LOGRADOURO POR GEORREFERENCIAMENTO'],
                          row['NÚMERO POR GEORREFERENCIAMENTO'], row["BAIRRO"])).add_to(mymap)

mymap
```

```python
mymap.save('mapa.html')
```
