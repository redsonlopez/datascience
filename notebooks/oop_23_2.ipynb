```python
from abc import ABC, abstractmethod

class ExtratorBase(ABC):
  def carregar_documento(self):
    """
    Esta função ira carregar o PDF ou imagem para o sistema
    """
    pass

  @abstractmethod
  def extrair_texto(self):
    raise NotImplementedError
```
