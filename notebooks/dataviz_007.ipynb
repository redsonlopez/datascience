<b>Valores aleatórios de uma produção com gráfico de linhas

```python
import matplotlib.pyplot as plt
import seaborn as sns
import random

# Loop para h de horas após cada número
horas = [f"{i}h" for i in range(8, 17)]

# Loops para geração de valores aleatórios para cada máquina
maquina_1 = [0,]
for i in range(8):
  numero = random.randint(350, 650)
  maquina_1.append(numero)

maquina_2 = [0,]
for i in range(8):
    numero = random.randint(550, 850)
    maquina_2.append(numero)

# Largura e altura do gráfico
plt.figure(figsize=(10,5))

# Obtém as coordenadas finais de cada linha
line1_coords = [(xi, yi) for xi, yi in zip(horas, maquina_1)]
line2_coords = [(xi, yi) for xi, yi in zip(horas, maquina_2)]

# Adiciona as legendas nos pontos de cada linha
for coord in line1_coords:
  plt.text(coord[0], coord[1], str(coord[1]), ha='right', va='bottom', color= 'gray')
for coord in line2_coords:
  plt.text(coord[0], coord[1], str(coord[1]), ha='right', va='bottom', color= 'gray')

# Configurações para às linhas do gráfico
sns.lineplot(x= horas, y= maquina_1, marker= 'o', color= 'blue', label= 'Maquína 1')
sns.lineplot(x= horas, y= maquina_2, marker= 's', color= 'navy', label= 'Maquína 2')

# Estilo e Tamanho da Fonte
sns.set_style('ticks')
plt.rcParams.update({'font.size': 8})

plt.ylim(0,900) # Ímagem de y

plt.suptitle('Produção',fontsize=13, color= 'navy', ha= 'right', va= 'top', y=1, x=0.22)
plt.xlabel('Horarios')
plt.ylabel('Quantidade Produzida')

plt.legend(loc= 'lower right') # Posição para Legendas das linhas

sns.despine() # Remove margem direita e superior do gráfico
```
