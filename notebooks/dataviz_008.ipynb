```python
import seaborn as sns
import pandas as pd
import numpy as np

iris = pd.read_csv('/content/drive/MyDrive/database/iris.xls')
iris.head(2)
```

```python
sns.relplot(x= 'SepalLengthCm', y= 'PetalLengthCm', data= iris, size= 'PetalWidthCm');
```

```python
sns.relplot(x= 'SepalLengthCm', y= 'PetalLengthCm', data= iris, hue= 'Species', style= 'Species', palette= 'Blues_d');
```

```python
sns.barplot(x= 'Species', y= 'SepalLengthCm', data= iris, errorbar= ('ci', 20));
```

```python
sns.barplot(y= 'Species', x= 'SepalLengthCm', data= iris, palette= 'Blues_d');
```

```python
iris['Local_Coleta'] = np.where(iris['Id']%2==0, 'Sp', 'Mg')
iris.head()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8,6)
```

```python
sns.barplot(x= 'Species', y= 'SepalLengthCm', data= iris, hue= 'Local_Coleta');
```

```python
sns.barplot(y= 'Species', x= 'SepalLengthCm', data= iris, palette= 'Blues_d');
```

```python
# TAREFA DE CASA
ax = sns.barplot(x='Species', y='SepalLengthCm', data=iris);

plt.figure(figsize=(10,8))
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height+.3,
            f'{round(p.get_height(),2)}',
            ha="center",
            rotation=45,
            fontsize=10)

ax.set_ylim(top=8) # limitador do eixo y
ax.set_title('Sepal Length por Espécie', fontsize=12); # titulo
ax.set_ylabel("Sepal Length Cm", fontsize=12); # rotulo do eixo y
ax.set_xlabel("Espécie", fontsize=12); # rotulo do eixo x
ax.tick_params(axis='x', labelsize=12, rotation=45) # marcacoes eixo x
ax.spines['top'].set_visible(False) # retirar linha do topo
ax.spines['right'].set_visible(False) # retirar linha lateral direita
```
