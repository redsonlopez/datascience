```python
from sqlalchemy import create_engine
import pandas as pd

class DataBase:
  def __init__(self):
    self._nome= "mysql+pymysql://mysql_user:mysql_password@mysql_host/mysql_db"
    self._con= create_engine(self._name)

  def query(self, q):
    """
    Execulta uma consulta ao banco de dados e retorna um DataFrame com o resultado.
    """
    df= pd.read_sql(q, con= self._con)
    return df

  def vendas_por_cliente(self, id_cliente):
    """
    Retorna um DataFrame com o historico de compras de um cliente.
    """

    query= f"""
    SELECT ID_CLIENTE, ID_PRODUTO, QUANTIDADE, DATA
    FROM VENDAS
    WHERE ID_CLIENTE= {id_cliente}
    """

    df= self.query(query)
    return df
```
