```python
def low_values_red(value):
  color = 'red' if value < 92 else 'black'
  return 'color: %s' %color # 'Color: red' ou 'Color: black'

fifa[['Overall', 'Potential']].head(10).style.applymap(low_values_red)
```

```python
fifa.head(10).style.applymap(low_values_red, subset= ['Overall', 'Potential'])
```

```python
def highlight_max(s):
  is_max = s == s.max() # Checa se s == s.max(), e depois guarda em is_max
  return ['background-color: yellow' if v else '' for v in is_max]

fifa.head(10).style.apply(highlight_max, subset= ['Age', 'Overall', 'Potential'])
```

```python
fifa[['Overall', 'Potential']].head(10).style.\
applymap(low_values_red).\
apply(highlight_max) # \ essa barra serve para quebra de linha
```

###<b>Estatística Descritiva

```python
fifa['Age'].describe(percentiles= [0.1, 0.2, 0.8, 0.99])
```

```python
fifa.agg({'Age': ['min', 'max', 'skew', 'median'], 'Overall': ['mean', 'min', 'max']})
```

```python
fifa_champs = fifa[fifa['Nationality'].isin(['Argentina', 'Brazil', 'Germany', 'France', 'Italy', 'England'])]

fifa_champs.groupby('Nationality')['Age', 'Overall'].mean().sort_values(by= 'Overall', ascending= False)
```

```python
import numpy as np
pd.pivot_table(fifa_champs, index= 'Nationality', aggfunc= np.mean)
```

```python
pd.pivot_table(fifa_champs, index=['Nationality'], values= ['Age', 'Overall', 'Strength'],aggfunc= np.mean)
```

```python
pd.pivot_table(fifa_champs, index=['Nationality', 'Position'], columns= ['Club'], values= ['Age', 'Overall', 'Strength'],aggfunc= np.mean)
```

```python
fifa_champs.Nationality.value_counts()
```
