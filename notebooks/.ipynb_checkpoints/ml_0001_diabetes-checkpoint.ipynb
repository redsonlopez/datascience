Linear Regression

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load datasets
diabetes_X, diabetes_y= datasets.load_diabetes(return_X_y= True)
# Select just one features
diabetes_X= diabetes_X[:, np.newaxis, 2]
# Splitting the database into training and test sets
diabetes_X_train= diabetes_X[:-20]
diabetes_X_test= diabetes_X[-20:]
# Division of outputs into training and testing
diabetes_y_train= diabetes_y[:-20]
diabetes_y_test= diabetes_y[-20:]
# Calls the Linear Regression function
regr= linear_model.LinearRegression()
# model training
regr.fit(diabetes_X_train, diabetes_y_train)
# Model prediction with test sets
diabetes_y_pred= regr.predict(diabetes_X_test)

# Coefficient
print("Coefficient: ", regr.coef_)
# Mean square error
print("Mean squared error: %.2f"%mean_squared_error(diabetes_y_test, diabetes_y_pred))
# Forecast coefficient. 1 is perfect
print("Forecast coefficient: %.2f" %r2_score(diabetes_y_test, diabetes_y_pred))
```

```python
# Output graph
sns.regplot(x= diabetes_X_test, y= diabetes_y_test, color= "Navy", line_kws={"color": "lightblue"});
plt.title("Linear Regression")
plt.xticks(())
plt.yticks(())
sns.despine()
```
