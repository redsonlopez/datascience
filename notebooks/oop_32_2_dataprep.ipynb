```python
import pandas as pd

class Vendas:
  def __init__(self):
    self.df_vendas= pd.read_csv("bases/t_vendas_csv")
    self.df_loja= pd.read_csv("bases/t_loja.csv")
    self.df_produto= pd.read_csv("bases/t_produto.csv")

  def remover_vendas_canceladas(self):
    self.df_vendas= self.df_vendas[self.df_vendas["status_vendas"] == 1]

  def tratar_categoria_faltantes(self):
    self.df_produto["cd_categoria"].fillna(-1, inplace= True)
    self.df_produto["nome_categoria"].fillna("outros", inplace= True)

  def combinar_bases(self):
    self.df_vendas= self.df_vendas.merge(self.df_loja)
    self.df_vendas= self.df_vendas.merge(self.df_produto)

  def tratar_dados(self):
    self.remover_vendas_canceladas()
    self.tratar_categoria_faltantes()
    self.combinar_bases()

  def gerar_relatorio_de_vendas(self):
    df= self.df_vendas.groupby(
        ["mes_faturamento", "nome_estado", "nome_categoria"]
    )["receita"].sum()

    return df
```
